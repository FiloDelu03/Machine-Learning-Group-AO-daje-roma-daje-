```python
import nbformat

# 1. Define the file names
file_eda = 'EDA.ipynb'
file_model = 'Model Car.ipynb'
output_name = 'Machine Learning Project Group AO.ipynb'

# 2. Read the notebooks
with open(file_eda, 'r', encoding='utf-8') as f:
    nb_eda = nbformat.read(f, as_version=4)

with open(file_model, 'r', encoding='utf-8') as f:
    nb_model = nbformat.read(f, as_version=4)

# 3. Merge cells (EDA first, then Model)
nb_eda.cells.extend(nb_model.cells)

# 4. Save the new merged notebook
with open(output_name, 'w', encoding='utf-8') as f:
    nbformat.write(nb_eda, f)

print(f"Success! '{output_name}' created safely.")
```
